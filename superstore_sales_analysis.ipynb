```python
# function to see the missing values

def find_missing_values(df,columns):
    missing_vals = {}
    print("Number of missing or NaN values for each column:")
    df_length = len(df)
    for column in columns:
        total_column_values = df[column].value_counts().sum()
        missing_vals[column] = df_length-total_column_values
    return missing_vals

missing_values = find_missing_values(df, columns=df.columns)
missing_values
```

# Exploratory data analysis

what is the overall sales trend?

```python
df['order_date'].min()
```

```python
df['order_date'].max()
```

```python
#change date format to month/year from timestamp above

df['month_year'] = df['order_date'].apply(lambda x: x.strftime('%Y-%m'))
```

```python
df['month_year']
```

```python
#grouping month_year

df_trend = df.groupby('month_year').sum()['sales'].reset_index()
```

```python
#plot df_trend vs sales

plt.figure(figsize=(15,6))
plt.plot(df_trend['month_year'],df_trend['sales'],c='red')
plt.xticks(rotation='vertical',size=8)
plt.show()
```
